```python
#Add repo path to the system path
from pathlib import Path
import os, sys
repo_path= Path.cwd().resolve()
while '.gitignore' not in os.listdir(repo_path): # while not in the root of the repo
    repo_path = repo_path.parent #go up one level
sys.path.insert(0,str(repo_path)) if str(repo_path) not in sys.path else None
```

```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import numpy as np
```

```python
probs_path = Path('experiments/inference/segmentation/data/predictions/full-size/multi-model_probs/MASK_106.nii.gz')
im_path = Path('data/challange_2023/Val/DATA/DATA_106.nrrd')
mask_path = Path('experiments/inference/segmentation/data/predictions/full-size/high-probs_no-lcc_limed_0.98_top-hat_0.0001_ideal-group/MASK_106.nii.gz')
```

```python
probs = sitk.ReadImage(str(probs_path))
probs = sitk.GetArrayFromImage(probs)
im = sitk.ReadImage(str(im_path))
im = sitk.GetArrayFromImage(im)
mask = sitk.ReadImage(str(mask_path))
mask = sitk.GetArrayFromImage(mask)
```

```python
slice_num = 193
# make figure fit the page

fig, ax = plt.subplots(3,1, figsize=(50,50))
ax[0].imshow(im[slice_num], cmap='gray')
ax[1].imshow(probs[slice_num], cmap='jet')
ax[2].imshow(mask[slice_num], cmap='gray')
# remove space between subplots
plt.subplots_adjust(wspace=0, hspace=0)
# remove axis
for a in ax:
    a.axis('off')
```
